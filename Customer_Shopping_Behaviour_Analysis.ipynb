```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#Create a column age-group
labels = ['Young Adult','Adult','Middle age','Senoir']
df['age_group'] = pd.qcut(df['age'], q=4, labels= labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create purchace frequency days

frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}

df['purchase_frequecny_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
